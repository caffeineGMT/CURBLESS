```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
place_name = "Union Rescue Mission"
place_address = "545 San Pedro St, Los Angeles, CA 90013"
graph = ox.graph_from_address(place_address, distance=3000)
```

```python
fig,ax=ox.plot_graph(graph)
```

```python
nodes,edges=ox.graph_to_gdfs(graph)
```

```python
buildings=ox.footprints.footprints_from_address(place_address,2000)
```

```python
fig,ax=plt.subplots()
edges.plot(ax=ax,linewidth=0.5, edgecolor='#BC8F8F')
buildings.plot(ax=ax,facecolor="khaki",alpha=0.7)
plt.tight_layout()
```
